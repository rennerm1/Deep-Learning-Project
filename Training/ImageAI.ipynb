# **Evaluation**

```
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/drive/My Drive/Object_Segmentation/models/detection_model-ex-026--loss-0003.503.h5") 
detector.setJsonPath("/content/drive/My Drive/Object_Segmentation/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/content/drive/My Drive/img_06677.jpg", output_image_path="/content/drive/My Drive/img_06677_detected.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
